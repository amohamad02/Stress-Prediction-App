```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv("data/train.csv.zip")
test = pd.read_csv("data/test.csv.zip")
```

```python
train.head()
```

```python
train.info()
```

```python
test.info()
```

```python
print(train.isnull().sum().sum())
print(test.isnull().sum().sum())
```

```python
train['condition'][train['condition']== "interruption"] = "time pressure"
test['condition'][test['condition']== "interruption"] = "time pressure"
```

```python
le = preprocessing.LabelEncoder()
le.fit(train['condition'])
train['condition'] = le.transform(train['condition'])
test['condition'] = le.transform(test['condition'])
```

```python
plt.figure(figsize=(12,10))
corr = train.corr()
sns.heatmap(corr, annot=False, cmap=plt.cm.Reds)
plt.show()
corr
```

```python
def feature_selection(correlation,threshold):
    selected_features = []
    for i in range(corr.shape[0]):
        if corr.iloc[i,35] > threshold:
            selected_features.append(train.iloc[:,i])
    return pd.DataFrame(selected_features).T
```

```python
reduced_train = feature_selection(corr,0.1)
```

```python
reduced_train
```

No stress = 0 and stress = 1

```python
reduced_train['condition'].unique()
```

```python
X_train = reduced_train.iloc[:,:-1]
y_train = reduced_train.iloc[:,-1]
```

```python
X_test = test[X_train.columns]
y_test = test['condition']
```

Defining the RandomForest Classifier

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
